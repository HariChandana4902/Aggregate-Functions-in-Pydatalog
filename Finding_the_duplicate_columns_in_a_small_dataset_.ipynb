We can find the duplicated columns or duplicated features in a particular dataset by transposing the dataset (converting rows into columns and columns into rows)
provided that dataset contains less than 5000 rows and 100 columns 
If we have more than thenumber specified above we will get runtime error

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/Santander.csv')
```

```python
data_t = data.T
```

```python
data_t
```

```python
data_t.duplicated().sum()  # used to find the number of duplicated columns n the dataset
```

```python
data_t[data_t.duplicated()] # used to visualize the duplicated columns
```

```python
array = data_t[data_t.duplicated()].index.values #inorder to capture the names of the columns alone
```

```python
array
```

```python
data.shape
```

```python
data_unique = data_t.drop_duplicates(keep = 'first').T
```

```python
data_unique.shape
```

```python
# So we can see that by dropping the duplicated columns, our datasize is reduced and it is easy to access and manipulate our data now
# We are reducing the feature space
```

```python
non_duplicated_features = [col for col in data.columns if col  in data_unique.columns]
non_duplicated_features 
```

```python
# The below code will list the names of the columns that are duplicated 
duplicated_features = [col for col in data.columns if col not in data_unique.columns]
duplicated_features
```

If the dataset contains more than 5000 rows and more than 100 columns, then we have to eliminate the columns manually without using any built-in function.
